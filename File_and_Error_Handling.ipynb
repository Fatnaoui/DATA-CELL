### moving the cursor
- seek(n): move the cursor to the nth byte from the beginning

```python
file.read()    # hello is missing because the cursor is at the 5th byte
```

### Smarter way of opening files
With the "with" statement, you get better syntax and exception handling. And also it will automatically close the file

```python
  with open("hello.txt",'r') as file:
    print(file.read(5))
    file.seek(0)
    print(file.read())
```

so, you should always use the with method to handle files in python

## Json files

let's first create a json file data.json
- {
  "name":"hamza",
  "age":21
}

```python
#let's read our data file
with open("data.json",'r') as file:
  print(file.read())
```

```python
#but if i wanna index it, i got an error
with open("data.json",'r') as file:
  print(file.read()["name"])
```

the problem is we don't know how our data is stored, we need to parse cad to understand it in dictionary forme. And to do that we have to import json

```python
import json
```

- load: from json file to python object
- loads: from json file(type string) to python object
- dump(odj,file): obj to json file
- dumps: obj(string) to json file

```python
# first way using load
with open("data.json",'r') as file:
  a=json.load(file)
  print(a["name"])
```

```python
# first way using loads
with open("data.json",'r') as file:
  a=file.read()   #in this case a is a string, when you read the file it gives you string
  d=json.loads(a)
  print(d["name"])
```

```python
dic={
    "name":"bikatchu",
    "age":12
}
```

```python
string=json.dumps(dic)
print(type(string))
```

```python
with open("data1.json",'w') as file:
  json.dump(dic,file)
```

```python
with open("data2.json",'w') as file:
  file.write(string)
```
