## Bar

```python
a=np.array([1,4,8])
plt.bar(a,[5,10,15],width=0.5,label="current year")
plt.bar(a+0.5,[15,10,5],width=0.5,label="next year")
plt.legend()
plt.show()
```

```python
#now let's label each bar
plt.style.use('dark_background')   # change the style
a=np.array([1,4,8])
plt.bar(a,[5,10,15],width=0.5,label="current year",tick_label=[2000,20002,2004])    #width/2 is how far the end of the bar from the center
plt.bar(a+0.5,[15,10,5],width=0.5,label="next year",color='green')
plt.xlabel("years")
plt.ylabel("productivity")
plt.title("pr yr")
plt.legend()
plt.show()
```

```python
#you can limit your axe by using
plt.style.use('dark_background')   # change the style
a=np.array([1,4,8])
plt.bar(a,[5,10,15],width=0.5,label="current year",tick_label=[2000,20002,2004])    #width/2 is how far the end of the bar from the center
plt.bar(a+0.5,[15,10,5],width=0.5,label="next year",color='green')
plt.xlabel("years")
plt.ylabel("productivity")
plt.title("pr yr")
plt.ylim(0,20)
plt.legend()
plt.show()
```

## Pie chart

```python
fields="GI","GEET","GIP"
per=[57,19,21]
plt.pie(per,labels=fields)
plt.show()
```

```python
fields="GI","GEET","GIP"
per=[57,19,21]
plt.pie(per,labels=fields,explode=(1,0,0))
plt.show()
```

```python
fields="GI","GEET","GIP"
per=[57,19,21]
plt.pie(per,labels=fields,explode=(0.1,0,0))
plt.show()
```

```python
# to get per
plt.pie(per,labels=fields,explode=(0.1,0,0),autopct='%1.1f%%')
plt.show()
```

```python
#plt.pie?  look for colors
plt.pie(per,labels=fields,explode=(0.1,0,0),colors=['blue','green','red'])
plt.show()
```

```python
# shadow
plt.pie(per,labels=fields,explode=(0.1,0,0),colors=['blue','green','red'],shadow=True)
plt.show()
```
